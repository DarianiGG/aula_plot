### Exemplo prático: Visualização de imagens de dígitos manuscritos
Vamos usar um dataset real para mostrar como a visualização pode ajudar na análise de dados.

```python
digits = load_digits()
print(digits.images[0])
```

```python
plt.figure(figsize=(6,6))
for i in range(4):
    ax = plt.subplot(2,2,i+1)
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f'Dígito {digits.target[i]}')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

#### O que podemos perceber ao visualizar as imagens?
- Cada pixel representa a intensidade do traço.
- Podemos identificar ruídos, padrões e diferenças entre os dígitos.
- Visualização é essencial para pré-processamento em visão computacional.

```python
# Agora, vamos calcular e visualizar a média de cada dígito para entender o padrão de cada classe.
mean_images = np.zeros((10,8,8))
for d in range(10):
    mean_images[d] = digits.images[digits.target==d].mean(axis=0)

fig, axs = plt.subplots(2,5, figsize=(10,5))
for i, ax in enumerate(axs.flatten()):
    ax.imshow(mean_images[i], cmap='inferno')
    ax.set_title(f'Média {i}')
    ax.axis('off')
plt.suptitle('Média de cada Dígito (8x8)')
plt.show()
```

## Exemplos práticos: Matplotlib
Vamos criar e customizar gráficos básicos e avançados.

```python
x = [1, 2, 3, 4, 5]
y = [10, 5, 15, 7, 12]
plt.figure(figsize=(6,4))
plt.plot(x, y, color='green', marker='o', linestyle='--')
plt.title('Exemplo de Gráfico de Linhas')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.grid(True)
plt.savefig('line_plot_example.png')
plt.show()
```

### Customizando gráficos: cores, legendas e rótulos

Esta seção tem como objetivo explicar o processo de customização de gráficos utilizando a biblioteca Matplotlib. A customização abrange a alteração de cores, a adição de legendas e a modificação de rótulos, permitindo que os gráficos sejam adaptados às necessidades específicas de apresentação de dados.

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 17, 35, 29]
plt.figure(figsize=(6,4))
bars = plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])
plt.title('Gráfico de Barras')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.show()
```

### Matplotlib avançado: subplots, estilos e anotações
Podemos criar múltiplos gráficos juntos, mudar o estilo visual e adicionar anotações para destacar pontos importantes.

```python
import numpy as np

# Dados para múltiplos subplots
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Aplicando estilo
plt.style.use("ggplot")

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, y1, color="purple")
axes[0].set_title("Seno")
axes[0].annotate(
    "Pico",
    xy=(np.pi / 2, 1),
    xytext=(np.pi / 2, 1.2),
    arrowprops=dict(arrowstyle="->", color="black"),
)

axes[1].plot(x, y2, color="teal")
axes[1].set_title("Cosseno")

plt.tight_layout()
plt.show()
```

## Exemplo com Seaborn
Gráfico de dispersão e histograma usando o dataset `tips`.

```python
tips = sns.load_dataset('tips')

plt.figure(figsize=(6,4))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
plt.title('Dispersão: Total Bill vs Tip')
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(data=tips, x='total_bill', bins=20, kde=True, color='skyblue')
plt.title('Histograma: Distribuição do Total Bill')
plt.show()
```

### Seaborn avançado: Boxplot, Violin Plot e Heatmap
Visualize distribuições, diferenças entre grupos e correlações.

```python
# Boxplot e Violin Plot
plt.figure(figsize=(6,4))
sns.boxplot(data=tips, x='day', y='total_bill', palette='Set2')
plt.title('Boxplot: Total Bill por Dia')
plt.show()

plt.figure(figsize=(6,4))
sns.violinplot(data=tips, x='day', y='total_bill', inner='quartile', palette='Set3')
plt.title('Violin Plot: Total Bill por Dia')
plt.show()

# Heatmap de correlação
corr = tips.select_dtypes(include='number').corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap de Correlação')
plt.show()
```

## Plotly avançado: Gráficos 3D e animações

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# Make figure with subplots
fig = make_subplots(rows=1, cols=2, specs=[[{"type": "bar"},
                                            {"type": "surface"}]])

# Add bar traces to subplot (1, 1)
fig.add_trace(go.Bar(y=[2, 1, 3]), row=1, col=1)
fig.add_trace(go.Bar(y=[3, 2, 1]), row=1, col=1)
fig.add_trace(go.Bar(y=[2.5, 2.5, 3.5]), row=1, col=1)

# Add surface trace to subplot (1, 2)
# Read data from a csv
z_data = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv")
fig.add_surface(z=z_data)

# Hide legend
fig.update_layout(
    showlegend=False,
    title_text="Default Theme",
    height=500,
    width=800,
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Create figure
fig = go.Figure()

# Add traces, one for each slider step
for step in np.arange(0, 5, 0.1):
    fig.add_trace(
        go.Scatter(
            visible=False,
            line=dict(color="#17C7CA", width=6),
            name="𝜈 = " + str(step),
            x=np.arange(0, 10, 0.01),
            y=np.sin(step * np.arange(0, 10, 0.01)),
        )
    )

# Make 10th trace visible
fig.data[10].visible = True

# Create and add slider
steps = []
for i in range(len(fig.data)):
    step = dict(
        method="update",
        args=[
            {"visible": [False] * len(fig.data)},
            {"title": "Slider switched to step: " + str(i)},
        ],  # layout attribute
    )
    step["args"][0]["visible"][i] = True  # Toggle i'th trace to "visible"
    steps.append(step)

sliders = [
    dict(active=10, currentvalue={"prefix": "Frequency: "}, pad={"t": 50}, steps=steps)
]

fig.update_layout(sliders=sliders)

fig.show()
```

## Daltonismo

Para pessoas daltonicas, é importante escolher paletas de cores que sejam acessíveis. Algumas dicas incluem:

- Usar contrastes altos entre as cores.
- Evitar combinações de cores problemáticas, como vermelho e verde.
- Utilizar texturas ou padrões para diferenciar elementos.

Existem ferramentas e bibliotecas que ajudam a criar visualizações mais inclusivas, como o Color Universal Design (CUD) e o uso de paletas de cores específicas no Matplotlib, para ajudar veja esse link [davidmathlogic](https://davidmathlogic.com/colorblind/).
